```python
import pandas as pd
```

```python
# Overview data
overview_2020_df = pd.read_csv("../data/overview_champions/overviews_champ_mapped/champMapped_2020_overview.csv", low_memory=False)
overview_2021_df = pd.read_csv("../data/overview_champions/overviews_champ_mapped/champMapped_2021_overview.csv", low_memory=False)
overview_2022_df = pd.read_csv("../data/overview_champions/overviews_champ_mapped/champMapped_2022_overview.csv", low_memory=False)
overview_2023_df = pd.read_csv("../data/overview_champions/overviews_champ_mapped/champMapped_2023_overview.csv", low_memory=False)
proximity_df = pd.read_csv("../data/proximity_data/proximity_data_all.csv", low_memory=False)
```

```python
overview_2020_df['gameid'] = overview_2020_df['gameid'].str.replace('/','_')
overview_2021_df['gameid'] = overview_2021_df['gameid'].str.replace('/','_')
overview_2022_df['gameid'] = overview_2022_df['gameid'].str.replace('/','_')
overview_2023_df['gameid'] = overview_2023_df['gameid'].str.replace('/','_')
overview_df = overview_2023_df
overview_df = overview_df.append(overview_2022_df, ignore_index=True)
overview_df = overview_df.append(overview_2021_df, ignore_index=True)
overview_df = overview_df.append(overview_2020_df, ignore_index=True)


proximity_df['gameId'] = proximity_df['gameId'].str.replace(':','_')
prox_columns = [ "tournament", "gameId", "team", "proximity_sum", "prox_score_1-5", "prox_score_5-10", "prox_score_10-15", "prox_score_15-20", "prox_score_20-25", "prox_score_25-30", "prox_score_30-35", "prox_score_35-40", "prox_score_40-45", "prox_score_45-50", "prox_score_50-55", "prox_score_55-60", "prox_score_60-65", "prox_score_65-70", "prox_score_70-75", "prox_score_75-80"]
proximity_df = proximity_df[prox_columns]
proximity_df = proximity_df.rename(columns={'gameId':'gameid', 'team': 'participantid'})
```

```python
overview_df
```

```python
merged_data = pd.merge(overview_df, proximity_df, on=['gameid', 'participantid'], how='inner')
merged_data
```
