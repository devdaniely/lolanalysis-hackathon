```python
import pandas as pd
from autogluon.tabular import TabularPredictor, TabularDataset
import autogluon.core as ag
from autogluon.common import space
from autogluon.tabular import TabularPredictor, TabularDataset
from sklearn.model_selection import train_test_split
```

```python
overview_data = TabularDataset(data="../data/all_data.csv")
df_train,df_test=train_test_split(overview_data,test_size=0.33,random_state=1)
test_data = df_test.drop(['result'],axis=1)
```

```python
predictor= TabularPredictor(label ='result').fit(train_data = df_train, verbosity = 2,presets='best_quality')
```

```python
predictor.fit_summary()
```

```python
predictor.leaderboard(df_train, silent=True)
```

```python
features = predictor.feature_importance(data=df_train)
```

```python
pd.set_option('display.max_rows', None)
features
```

```python
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred,columns=['stroke'])
y_pred #print the DataFrame
```
